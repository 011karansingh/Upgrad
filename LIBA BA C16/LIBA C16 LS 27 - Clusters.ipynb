```python
df = pd.read_csv('Mall_Customers.csv')
```

```python
df
```

```python
sns.lmplot(data=df, x='Annual Income (k$)',y='Spending Score (1-100)',fit_reg=False)
```

```python
x = df.iloc[:,3:5].values
```

```python
x
```

```python
from sklearn.cluster import KMeans
wcss = []
for  i in range(1,11):
    kmeans = KMeans(n_clusters=i, init='k-means++')
    kmeans.fit(x)
    wcss.append(kmeans.inertia_)
```

```python
plt.plot(range(1,11), wcss)
```

```python
kmeans = KMeans(n_clusters=5, init='k-means++',random_state=10)
y_pred = kmeans.fit_predict(x)
```

```python
y_pred
```

```python
output = pd.concat([df,pd.DataFrame(y_pred)],axis=1)
```

```python
output.rename(columns={0:'Clusters'},inplace=True)
```

```python
output
4 - LSLSS
2 - LSHSS
3 - HSHSS
0 - HSLSS
1 - ASASS
```

```python
output['Clusters'] = output.Clusters.map({0:'HSLSS',1:'ASASS',2:'LSHSS',3:'HSHSS',4:'LSLSS'})
```

```python
output
```

```python
sns.lmplot(data=output, x='Annual Income (k$)',y='Spending Score (1-100)',hue='Clusters',fit_reg=False)
```

```python
from sklearn.metrics import silhouette_score
silhouette_score(x,y_pred)
```

```python
output[output['Clusters']=='HSHSS'].to_csv('Output_HSHSS.csv',index=False)
```
